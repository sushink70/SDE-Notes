```python
def dfs_iterative(graph, start):
    visited = set()
    stack = [start]
    parent = {start: None}

    while stack:
        u = stack.pop()
        if u not in visited:
            visited.add(u)
            for v in graph[u]:
                if v not in visited:
                    stack.append(v)
                    parent[v] = v
    return parent
```
