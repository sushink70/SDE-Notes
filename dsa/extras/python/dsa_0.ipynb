```go
arr = [5, 6, 2, 6, 9, 1]
new_arr = sorted(arr)
print(new_arr)  # [1, 2, 5, 6, 6, 9]
```

```go
ne_arr = [5, 6, 2, 6, 9, 1]
ne_arr.sort()
print(ne_arr)  # [1, 2, 5, 6, 6, 9]
print(ne_arr.sort()) # None

# Explanation:
# .sort() modifies the list IN-PLACE and returns None
# It doesn't return a new list, it changes the original

# Correct usage:
arr1 = [5, 6, 2, 6, 9, 1]
arr1.sort()  # Modifies arr1
print("After .sort():", arr1)  # [1, 2, 5, 6, 6, 9]

# vs sorted() - returns a new list
arr2 = [5, 6, 2, 6, 9, 1]
result = sorted(arr2)  # Returns new list, arr2 unchanged
print("sorted() result:", result)  # [1, 2, 5, 6, 6, 9]
print("Original arr2:", arr2)     # [5, 6, 2, 6, 9, 1]
```

```go
import "fmt"

func main() {
	arr := []int{2, 4, 6, 8, 10}
	s := len(arr)
	fmt.Println(s)

	ss := "Hwwwwwwwww"
	l := len(ss)
	fmt.Println(l)

	dictt := map[string]int{"a": 1, "v": 4, "f": 0}
	fmt.Println(len(dictt))

	arr = []int{2, 4, 6, 8, 10}
	for i := len(arr) - 1; i >= 0; i-- {
		fmt.Println(arr[i])
	}
	
	arr = []int{2, 4, 6, 8, 10}
	for index, value := range arr {
		fmt.Println(index, value)
	}

	for index, value := range arr {
		fmt.Println(index+1, value)
	}

	hashmap := make(map[int]int)
	for key, value := range arr {
		hashmap[key] = value // means key is index, value is arr[index]
	}
	fmt.Println(hashmap) // map[0:2 1:4 2:6 3:8 4:10]

	names := []string{"Alice", "Bob", "Charlie"}
	ages := []int{25, 30, 35}
	cities := []string{"NYC", "LA", "Chicago"}

	for i := range names {
		fmt.Printf("%s is %d years old and lives in %s.\n", names[i], ages[i], cities[i])
	}

	dict := make(map[string]int)
	for i := range names {
		dict[names[i]] = ages[i]
	}
	fmt.Println(dict)

	arr = []int{5, 6, 2, 6, 9, 1}
	newArr := make([]int, len(arr))
	copy(newArr, arr)
	// Sorting newArr
	for i := 0; i < len(newArr)-1; i++ {
		for j := i + 1; j < len(newArr); j++ {
			if newArr[i] > newArr[j] {
				newArr[i], newArr[j] = newArr[j], newArr[i]
			}
		}
	}
	fmt.Println(newArr) // [1, 2, 5, 6, 6, 9]

	neArr := []int{5, 6, 2, 6, 9, 1}
	// Sorting neArr in place
	for i := 0; i < len(neArr)-1; i++ {
		for j := i + 1; j < len(neArr); j++ {
			if neArr[i] > neArr[j] {
				neArr[i], neArr[j] = neArr[j], neArr[i]
			}
		}
	}
	fmt.Println(neArr) // [1, 2, 5, 6, 6, 9]
}
```
